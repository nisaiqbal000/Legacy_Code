```python
import os
import datetime

# File names for storage
PATIENTS_FILE = 'patients.txt'
DOCTORS_FILE = 'doctors.txt'
APPOINTMENTS_FILE = 'appointments.txt'

def load_data(filename):
    if not os.path.exists(filename):
        return []
    with open(filename, 'r') as file:
        return [line.strip() for line in file.readlines()]

def save_data(filename, data_list):
    with open(filename, 'w') as file:
        for item in data_list:
            file.write(f"{item}\n")

def add_patient():
    pid = input("Enter Patient ID: ")
    name = input("Enter Patient Name: ")
    age = input("Enter Age: ")
    gender = input("Enter Gender: ")
    contact = input("Enter Contact Number: ")
    patient_record = f"{pid},{name},{age},{gender},{contact}"
    patients = load_data(PATIENTS_FILE)
    patients.append(patient_record)
    save_data(PATIENTS_FILE, patients)
    print("Patient record added successfully.")

def view_patients():
    patients = load_data(PATIENTS_FILE)
    print("\n--- Patient Records ---")
    for p in patients:
        pid, name, age, gender, contact = p.split(",")
        print(f"ID: {pid}, Name: {name}, Age: {age}, Gender: {gender}, Contact: {contact}")

def add_doctor():
    did = input("Enter Doctor ID: ")
    name = input("Enter Doctor Name: ")
    specialization = input("Enter Specialization: ")
    contact = input("Enter Contact Number: ")
    doctor_record = f"{did},{name},{specialization},{contact}"
    doctors = load_data(DOCTORS_FILE)
    doctors.append(doctor_record)
    save_data(DOCTORS_FILE, doctors)
    print("Doctor record added successfully.")

def view_doctors():
    doctors = load_data(DOCTORS_FILE)
    print("\n--- Doctor Records ---")
    for d in doctors:
        did, name, specialization, contact = d.split(",")
        print(f"ID: {did}, Name: {name}, Specialization: {specialization}, Contact: {contact}")

def book_appointment():
    patient_id = input("Enter Patient ID: ")
    doctor_id = input("Enter Doctor ID: ")
    date = input("Enter Appointment Date (YYYY-MM-DD): ")
    time = input("Enter Appointment Time (HH:MM): ")
    appointment_id = f"A{datetime.datetime.now().strftime('%Y%m%d%H%M%S')}"
    appointment_record = f"{appointment_id},{patient_id},{doctor_id},{date},{time}"
    appointments = load_data(APPOINTMENTS_FILE)
    appointments.append(appointment_record)
    save_data(APPOINTMENTS_FILE, appointments)
    print("Appointment booked successfully.")

def view_appointments():
    appointments = load_data(APPOINTMENTS_FILE)
    print("\n--- Appointments ---")
    for a in appointments:
        aid, pid, did, date, time = a.split(",")
        print(f"Appointment ID: {aid}, Patient ID: {pid}, Doctor ID: {did}, Date: {date}, Time: {time}")

def main_menu():
    while True:
        print("\n--- Hospital Management System ---")
        print("1. Add Patient")
        print("2. View Patients")
        print("3. Add Doctor")
        print("4. View Doctors")
        print("5. Book Appointment")
        print("6. View Appointments")
        print("7. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            add_patient()
        elif choice == '2':
            view_patients()
        elif choice == '3':
            add_doctor()
        elif choice == '4':
            view_doctors()
        elif choice == '5':
            book_appointment()
        elif choice == '6':
            view_appointments()
        elif choice == '7':
            print("Exiting system. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

# Run the program
if __name__ == "__main__":
    main_menu()
```
